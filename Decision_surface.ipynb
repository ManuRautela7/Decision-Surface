```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets,neighbors
```

```python
from matplotlib.colors import ListedColormap

def Knn_comparision(data,n_neighbors):
    
    #X contains the x and y values for the points
    X=data[:, :2]
    
    #Y contains the values of the class_label
    Y=data[:,2]
    
    #grid cell size
    h=0.2
    
    #assigning colors
    cmap_light=ListedColormap(['#FFAAAA',  '#AAAAFF'])
    cmap_bold=ListedColormap(['#FF0000', '#0000FF'])
    
    #core classifier KNN
    clf = neighbors.KNeighborsClassifier(n_neighbors)
    clf.fit(X,Y)
    
    #assigning minimum and maximum values to x_min and x_max
    x_min,x_max=X[:,0].min()-1,X[:,0].max()+1
    
    #assigning minimum and maximum values to y_min and y_max
    y_min,y_max=X[:,1].min()-1,X[:,1].max()+1
    
    # np.arange(x_min,x_max,h) creates an array with lower limit as x_min and upper limit as x_max and all other values at an interval of h
    #creating the meshgrid 
    xx,yy=np.meshgrid(np.arange(x_min,x_max,h),np.arange(y_min,y_max,h))
    
    # xx and yy will contain the same array multiple times asa lists within lists, so that we can perform np.c to combine x and y to create a point. 
    
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    
     # xx.ravel() will give a flatten array

    # np.c_ : Translates slice objects to concatenation along the second axis.
    # > np.c_[np.array([1,2,3]), np.array([4,5,6])]
    # > array([[1, 4],
    #          [2, 5],
    #          [3, 6]])   (source: np.c_ documentation)
    

    Z=Z.reshape(xx.shape)
    plt.figure()
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light)
   
    # scatter plot of with given points
    plt.scatter(X[:, 0], X[:, 1], c=Y, cmap=cmap_bold)
    
    #defining scale on both axises
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
   
    # set the title
    plt.title('K value = '+str(n_neighbors))
    

    plt.show()
```

```python
data1 = np.genfromtxt('6.overlap.csv', delimiter=',')
data1.shape
Knn_comparision(data1,1)
Knn_comparision(data1,5)
Knn_comparision(data1,10)
Knn_comparision(data1,15)
Knn_comparision(data1,20)
```
